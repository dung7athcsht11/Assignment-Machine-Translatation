```python
# CELL 5: Hàm chia ALT corpus 80/20 (chạy 1 lần thôi, lưu file mới vào Drive)
def split_alt_corpus():
    alt_path = "/content/drive/MyDrive/ALTCorpus"
    en_file = f"{alt_path}/data_en.txt"
    vi_file = f"{alt_path}/data_vi.txt"

    # Đọc lines
    en_lines = [line.strip() for line in open(en_file, encoding='utf-8')]
    vi_lines = [line.strip() for line in open(vi_file, encoding='utf-8')]

    n = len(en_lines)
    split_idx = int(0.8 * n)

    train_en = en_lines[:split_idx]
    train_vi = vi_lines[:split_idx]
    val_en = en_lines[split_idx:]
    val_vi = vi_lines[split_idx:]

    # Lưu file mới
    with open(f"{alt_path}/data_en_train.txt", 'w', encoding='utf-8') as f:
        f.write('\n'.join(train_en))
    with open(f"{alt_path}/data_vi_train.txt", 'w', encoding='utf-8') as f:
        f.write('\n'.join(train_vi))
    with open(f"{alt_path}/data_en_val.txt", 'w', encoding='utf-8') as f:
        f.write('\n'.join(val_en))
    with open(f"{alt_path}/data_vi_val.txt", 'w', encoding='utf-8') as f:
        f.write('\n'.join(val_vi))

    print(f"Đã chia ALT: {split_idx} train / {n-split_idx} val | Tổng {n} câu")

split_alt_corpus()
```

```python

class TranslationDataset(Dataset):
    def __init__(self, src_file, tgt_file):
        self.src_lines = [line.strip() for line in open(src_file, encoding="utf-8")]
        self.tgt_lines = [line.strip() for line in open(tgt_file, encoding="utf-8")]
        assert len(self.src_lines) == len(self.tgt_lines), f"Số dòng không khớp ở {src_file}"

    def __len__(self):
        return len(self.src_lines)

    def __getitem__(self, idx):
        src = [2] + sp.encode(self.src_lines[idx], out_type=int) + [3]   # bos + tokens + eos
        tgt = [2] + sp.encode(self.tgt_lines[idx], out_type=int) + [3]
        return torch.tensor(src), torch.tensor(tgt)

def collate_fn(batch):
    src_batch, tgt_batch = zip(*batch)
    src_batch = nn.utils.rnn.pad_sequence(src_batch, padding_value=1, batch_first=True)
    tgt_batch = nn.utils.rnn.pad_sequence(tgt_batch, padding_value=1, batch_first=True)
    return src_batch, tgt_batch
```

```python
# CELL MODEL – PHIÊN BẢN CUỐI CÙNG, ĐÃ FIX HẾT TẤT CẢ LỖI
class PositionalEncoding(nn.Module):
    def __init__(self, d_model=512, max_len=1024):
        super().__init__()
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len).unsqueeze(1).float()
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        self.register_buffer('pe', pe.unsqueeze(0))

    def forward(self, x):
        return x + self.pe[:, :x.size(1)]

class RMSNorm(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.weight = nn.Parameter(torch.ones(dim))
    def forward(self, x):
        return x * torch.rsqrt(x.pow(2).mean(-1, keepdim=True) + 1e-6) * self.weight

class NMTModel(nn.Module):
    def __init__(self, variant="re"):           # "re" = Re-Transformer, "original" = Classic
        super().__init__()
        d_model = 512
        nhead   = 8
        num_layers = 6

        self.src_emb = nn.Embedding(vocab_size, d_model)
        self.tgt_emb = nn.Embedding(vocab_size, d_model)
        self.pos     = PositionalEncoding(d_model)

        # Activation + Norm theo variant
        activation = "relu" if variant == "re" else "gelu"
        self.final_norm = RMSNorm(d_model) if variant == "re" else nn.LayerNorm(d_model)

        # Encoder
        enc_layer = nn.TransformerEncoderLayer(
            d_model=d_model, nhead=nhead, dim_feedforward=2048,
            dropout=0.1, activation=activation, batch_first=True
        )
        self.encoder = nn.TransformerEncoder(enc_layer, num_layers)

        # Decoder
        dec_layer = nn.TransformerDecoderLayer(
            d_model=d_model, nhead=nhead, dim_feedforward=2048,
            dropout=0.1, activation=activation, batch_first=True
        )
        self.decoder = nn.TransformerDecoder(dec_layer, num_layers)

        self.out = nn.Linear(d_model, vocab_size)

        # Re-Transformer style initialization
        if variant == "re":
            for p in self.parameters():
                if p.dim() > 1:
                    nn.init.xavier_uniform_(p)

    def forward(self, src, tgt):
        """
        src: [batch_size, src_len]
        tgt: [batch_size, tgt_len]   ← đã có <bos> và sẽ có <eos> ở cuối
        """
        src_len = src.size(1)
        tgt_len = tgt.size(1)

        # 1. Encoder self-attention mask: [src_len, src_len]
        src_mask = nn.Transformer.generate_square_subsequent_mask(src_len).to(device)

        # 2. Decoder self-attention mask: [tgt_len, tgt_len]
        tgt_mask = nn.Transformer.generate_square_subsequent_mask(tgt_len).to(device)

        # 3. Cross-attention mask (memory_mask): phải là [tgt_len, src_len]
        #    Ở đây ta cho phép decoder nhìn thấy toàn bộ encoder → để trống (None cũng được)
        memory_mask = None

        # Embedding + positional encoding
        src = self.pos(self.src_emb(src) * math.sqrt(512.0))
        tgt = self.pos(self.tgt_emb(tgt) * math.sqrt(512.0))

        # Encoder
        memory = self.encoder(src, mask=src_mask)

        # Decoder
        output = self.decoder(
            tgt, memory,
            tgt_mask=tgt_mask,
            memory_mask=memory_mask
        )

        # Final norm + linear
        return self.out(self.final_norm(output))
```

```python

def train_corpus(name, src_train, tgt_train, src_val, tgt_val, epochs=6):
    print(f"\n{'='*30} {name.upper()} {'='*30}")
    train_ds = TranslationDataset(src_train, tgt_train)
    val_ds   = TranslationDataset(src_val,   tgt_val)

    # Giảm batch_size xuống 32 + tích luỹ gradient (effective batch = 128)
    train_loader = DataLoader(train_ds, batch_size=32, shuffle=True,  collate_fn=collate, num_workers=2, pin_memory=True)
    val_loader   = DataLoader(val_ds,   batch_size=32, shuffle=False, collate_fn=collate, num_workers=2, pin_memory=True)

    results = []
    for variant in ["original", "re"]:
        print(f"\n{variant.upper()} TRANSFORMER")
        model = NMTModel(variant=variant).to(device)

        # Mixed precision (tiết kiệm VRAM ~50%)
        scaler = torch.amp.GradScaler()

        opt = optim.AdamW(model.parameters(), lr=3e-4, betas=(0.9,0.98))
        crit = nn.CrossEntropyLoss(ignore_index=1)

        best_val = 99.0
        start = time.time()

        for epoch in range(1, epochs+1):
            model.train()
            total_loss = 0
            opt.zero_grad()

            for i, (s, t) in enumerate(train_loader):
                s, t = s.to(device, non_blocking=True), t.to(device, non_blocking=True)

                with torch.amp.autocast(device_type='cuda'):   # ← magic line
                    out = model(s, t[:, :-1])
                    loss = crit(out.view(-1, vocab_size), t[:, 1:].reshape(-1))
                    loss = loss / 4   # gradient accumulation steps = 4 → effective batch 128

                scaler.scale(loss).backward()

                if (i + 1) % 4 == 0 or (i + 1) == len(train_loader):
                    scaler.unscale_(opt)
                    nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    scaler.step(opt)
                    scaler.update()
                    opt.zero_grad()

                total_loss += loss.item() * 4

            # Validation (cũng dùng mixed precision)
            model.eval()
            val_loss = 0
            with torch.no_grad():
                for s, t in val_loader:
                    s, t = s.to(device), t.to(device)
                    with torch.amp.autocast(device_type='cuda'):
                        out = model(s, t[:, :-1])
                        val_loss += crit(out.view(-1, vocab_size), t[:, 1:].reshape(-1)).item()
            val_loss /= len(val_loader)

            print(f"Epoch {epoch:02d} → Train {total_loss/len(train_loader):.4f} | Val {val_loss:.4f}")

            if val_loss < best_val:
                best_val = val_loss
                torch.save(model.state_dict(), f"/content/drive/MyDrive/{name}_{variant}_best.pt")

        duration = time.time() - start
        results.append([f"{'Re' if variant=='re' else 'Classic'} Transformer", f"{duration/60:.1f} phút", f"{best_val:.4f}"])

    print(f"\nKẾT QUẢ {name.upper()}")
    print("-"*65)
    print(f"{'Model':<22} {'Thời gian':<15} {'Best Val Loss'}")
    print("-"*65)
    for r in results:
        print(f"{r[0]:<22} {r[1]:<15} {r[2]}")
    print("="*65)
```
